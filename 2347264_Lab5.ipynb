```python
import kagglehub
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import glob as gb
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import tensorflow as tf

# Step 1: Downloading the dataset using kagglehub
print("Downloading dataset...")
dataset_path = kagglehub.dataset_download('puneet6060/intel-image-classification')

# Step 2: Setting up paths for data folders
train_path = os.path.join(dataset_path, 'seg_train', 'seg_train')
test_path = os.path.join(dataset_path, 'seg_test', 'seg_test')

# Step 3: Define category labels
categories = {
    'buildings': 0,
    'forest': 1,
    'glacier': 2,
    'mountain': 3,
    'sea': 4,
    'street': 5
}

# Step 4: Function to load and preprocess images

def load_images(path, image_size=100):
    """
    Load and preprocess images with explicit data type handling
    """
    images = []
    labels = []

    for category in os.listdir(path):
        files = gb.glob(os.path.join(path, category, '*.jpg'))
        print(f'Loading {len(files)} images from {category}')

        for file in files:
            image = cv2.imread(file)
            if image is not None:
                image = cv2.resize(image, (image_size, image_size))
                # Ensure image is uint8
                image = image.astype(np.uint8)
                images.append(image)
                labels.append(categories[category])

    return np.array(images, dtype=np.uint8), np.array(labels)

def visualize_samples(X, y, categories, num_samples=5):
    """
    Visualize sample images with proper type conversion
    """
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))
    for i in range(num_samples):
        # Random samples from two different classes
        idx1 = np.random.choice(np.where(y == i)[0])
        idx2 = np.random.choice(np.where(y == (i+1)%6)[0])

        # Convert to uint8 before color conversion
        img1 = (X[idx1] * 255).astype(np.uint8) if X[idx1].dtype != np.uint8 else X[idx1]
        img2 = (X[idx2] * 255).astype(np.uint8) if X[idx2].dtype != np.uint8 else X[idx2]

        axes[0, i].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
        axes[0, i].set_title(list(categories.keys())[y[idx1]])
        axes[0, i].axis('off')

        axes[1, i].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
        axes[1, i].set_title(list(categories.keys())[y[idx2]])
        axes[1, i].axis('off')

    plt.tight_layout()
    plt.show()

# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, categories):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=categories.keys(),
                yticklabels=categories.keys())
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

# Step 5: Load training and test data
print("\nLoading training data...")
X_train, y_train = load_images(train_path)
print("\nLoading test data...")
X_test, y_test = load_images(test_path)

# Step 6: Normalize pixel values
# X_train = X_train / 255.0
# X_test = X_test / 255.0
X_train = (X_train / 255.0).astype(np.float32)
X_test = (X_test / 255.0).astype(np.float32)


# Step 7: Split training data to create a validation set
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42
)

# Step 8: Visualize sample images
print("\nVisualizing sample images...")
visualize_samples(X_train, y_train, categories)

# Step 9: Set up data augmentation
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    fill_mode='nearest'
)

# Step 10: Create the enhanced CNN model
model = Sequential([
    # First Convolutional Block
    Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(100, 100, 3)),
    BatchNormalization(),
    Conv2D(32, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Second Convolutional Block
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Third Convolutional Block
    Conv2D(128, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    Conv2D(128, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Dense layers
    Flatten(),
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(6, activation='softmax')
])

# Step 11: Set up learning rate scheduling
initial_learning_rate = 0.001
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=1000,
    decay_rate=0.9
)
optimizer = Adam(learning_rate=lr_schedule)

# Step 12: Compile the model
model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
print("\nModel Architecture:")
model.summary()

# Step 13: Set up training callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=3
    )
]

# Step 14: Train the model
print("\nTraining the model...")
history = model.fit(
    train_datagen.flow(X_train, y_train, batch_size=32),
    epochs=10,
    validation_data=(X_val, y_val),
    callbacks=callbacks
)

# Step 15: Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'\nTest accuracy: {test_accuracy:.4f}')

# Step 16: Plot training results
plt.figure(figsize=(12, 4))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Step 17: Generate and plot confusion matrix
print("\nGenerating confusion matrix...")
y_pred = np.argmax(model.predict(X_test), axis=1)
plot_confusion_matrix(y_test, y_pred, categories)

# Step 18: Function to make predictions on new images
def predict_image(image_path):
    # Read and preprocess the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, (100, 100))
    image = image / 255.0
    image = np.expand_dims(image, axis=0)

    # Make prediction
    prediction = model.predict(image)
    class_idx = np.argmax(prediction[0])

    # Get category name
    category_name = list(categories.keys())[list(categories.values()).index(class_idx)]
    confidence = prediction[0][class_idx]

    # Display result
    plt.imshow(cv2.cvtColor(image[0], cv2.COLOR_BGR2RGB))
    plt.title(f'Prediction: {category_name}\nConfidence: {confidence:.2%}')
    plt.axis('off')
    plt.show()

print("\nModel training and evaluation complete!")
print("You can now use the predict_image() function to classify new images.")

# Example usage of predict_image function:
# predict_image(train_path)
```
